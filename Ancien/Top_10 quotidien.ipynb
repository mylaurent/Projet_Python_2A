#### Création de dico_score

On crée un dico avec les caractéristiques intéressantes

```python
liste_df = [df_decembre,df_novembre,df_octobre,df_septembre,df_aout,df_juillet,df_juin,df_mai,df_avril,df_mars]
dico_carac=dict()

for df_moi in liste_df:
    for i in range(sortby(df_moi).shape[0]):
        if sortby(df_moi)['Title'][i] not in dico_carac:
            dico_carac[sortby(df_moi)['Title'][i]]=[sortby(df_moi)['Type'][i],sortby(df_moi)['Netflix Excl.'][i],sortby(df_moi)['Viewer- Ship Score'][i]]

dico_carac
```

On convertit le score en entier

```python
for cle in dico_carac.keys():
    u=dico_carac[cle][2]
    dico_carac[cle][2] = int(u)
dico_carac
```

On crée une fonction qui crée un dataframe à partir du dico

```python
def dict_to_df(dico,columns_titles,sortby='Score'):
    df = pd.DataFrame.from_dict(dico,orient='index')
    df.columns=columns_titles
    return df.sort_values(sortby,ascending=False)
```

On crée le df et on le stock dans un fichier csv.

```python
df_carac=dict_to_df(dico_carac,['Type','Netflix Exclusive','Score'])
df_carac['Netflix Exclusive']=df_carac['Netflix Exclusive'].fillna('No')
gfg_csv_data = df_carac.to_csv('df_carac.csv', index = True) 
```
