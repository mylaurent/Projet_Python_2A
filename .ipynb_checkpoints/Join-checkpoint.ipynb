### Join score et df_tvshow et df_movie -> très peu de valeurs nn nulles de Score

```python
df_tvshow= pd.read_csv(r'C:\Users\Utilisateur\Documents\GitHub\Projet_Python_2A\df_tvshow.csv',sep=',')
df_movie= pd.read_csv(r'C:\Users\Utilisateur\Documents\GitHub\Projet_Python_2A\df_movie.csv',sep=',')
```

```python
df_type_score= pd.read_csv(r'C:\Users\Utilisateur\Documents\GitHub\Projet_Python_2A\type_score.csv')
df_type_score.columns=['Title','Type','Score']
```

```python
OGTV_join_score=df_tvshow.set_index('title').join(df_type_score.set_index('Title'))
OGTV_join_score.sort_values('Score',ascending='False') #1969 row
OGTV_join_score['Score'].notna().sum()
gfg_csv_data = OGTV_join_score.to_csv('OGTV_join_score.csv', index = True) 
```

```python
OGM_join_score=df_movie.set_index('title').join(df_type_score.set_index('Title'))
OGM_join_score.sort_values('Score',ascending='False')
OGM_join_score['Score'].notna().sum()
gfg_csv_data = OGM_join_score.to_csv('OGM_join_score.csv', index = True) 
```
